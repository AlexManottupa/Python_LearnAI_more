```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(2,9,10)
x
```

```python
y=x**2
y
```

```python
plt.title('Este es un grafico')
plt.plot(x,y, 'blue' )
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y, 'g')

plt.subplot(2,1,2)
plt.plot(x,y, 'blue')
```

```python
figura=plt.figure()
grafico = figura.add_axes([0.1,0.1,0.5,0.5]) 
grafico.plot(x,y, 'green')
```

```python
figura=plt.figure()
grafico = figura.add_axes([0.1,0.1,1,1]) 
grafico.plot(x,y, 'g')
 

grafico2=figura.add_axes([0.2,0.4,0.3,0.3])
grafico2.plot(x,y,'r') 
```

```python
figura, graficos = plt.subplots(nrows=2,ncols=2)
graficos[0][0].plot(x,y,'blue')
graficos[0][1].plot(x,y,'red')
graficos[1][0].plot(x,y,'green')
graficos[1][1].plot(x,y,'yellow')

graficos[0][0].set_title('Azul')
graficos[0][1].set_title('Rojo')
graficos[1][0].set_title('Verde')
graficos[1][1].set_title('Amarillo')
```

```python
figure = plt.figure(figsize=(4.8))
grafico=figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,'r')
```

```python
figura = plt.figure()
graficos=figura.add_axes([0.1,0.1,1,1])
graficos.plot(x,y**2, label='Al cuadrado')

graficos.plot(x,y**3, label='Al cubo')
graficos.legend(loc=0)
```

```python
figura=plt.figure()
grafico=figura.add_axes([0,0,1,1])
grafico.plot(x,y,'#1af3c9')
```

```python
figura=plt.figure()
grafico=figura.add_axes([0,0,1,1])
grafico.plot(x,y,color='blue' ,linewidth=15,alpha=0.9,linestyle='dotted')
```

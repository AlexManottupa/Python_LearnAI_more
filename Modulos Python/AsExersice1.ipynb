```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.DataFrame()
df["Age"] = [18, 20, 22, 25, 28, 34] 
df["Salary"] = [1100, 1400, 1800, 2500, 3500, 4600]
```

```python
df
```

```python
plt.scatter(df["Age"], [df["Salary"]])
```

```python
X = df[["Age"]]
```

```python
y = df["Salary"]
```

```python
modelo = linear_model.LinearRegression()
```

```python
modelo.fit(X, y)
```

```python
modelo.predict([[26]])
```

```python
plt.scatter(df["Age"], df["Salary"])
plt.plot(df["Age"], modelo.predict(X))
```

```python
import nltk
nltk.download()
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Tokenizar texto

#leer texto desde html



from bs4 import BeautifulSoup

import urllib.request



url = 'https://librefinanciero.com/'

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}



req = urllib.request.Request(url, headers=headers)



try:

  response = urllib.request.urlopen(req)

  html = response.read()

  # Try using a different parser to see if the issue is specific to html5lib

  soup = BeautifulSoup(html, "html.parser") # Switch to html.parser or lxml

  text = soup.get_text(strip=True)

  print(text)

except Exception as e:

  print(f'An unexpected error occurred: {e}')
```

```python
#Crea un tokens de palabras
#Tokens en palabras
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text, 'english' )
tokens = [maria.lower() for maria in tokens if maria.isalpha()] # para remover los signos de puntuacion
tokens
```

```python
freq = nltk.FreqDist(tokens)
for key, val in freq.items():
    {
        print(str(key) +' : ' +str(val))
    }
```

```python
#Visualizar tokens

sns.set()
freq.plot(30, cumulative=False)
```

```python
#Eliminar palabras de parada
from nltk.corpus import stopwords
clean_tokens = tokens[:]
for token in tokens:
    if tokens in stopwords.words('spanish'):
        clean_tokens.remove(token)
```

```python
clean_tokens
```

```python
#verificar frecuecia de palabras 
freq_clean = nltk.FreqDist(clean_tokens)
for key, val in freq_clean.items():
    print(str(key) + ' : ' + str(val))
```

```python
#Obtener antonimos
from nltk.corpus import wordnet
sinonimos = []

for syn in wordnet.synsets('investment'):
    for lemma in syn.lemmas():
        sinonimos.append(lemma.name())


sinonimos
```

```python
#Obtener antonimos
from nltk.corpus import wordnet
antonimos = []

for syn in wordnet.synsets('good'):
    for l in syn.lemmas():
        if l.antonyms():
            antonimos.append(l.antonyms()[0].name())
                         

antonimos
```

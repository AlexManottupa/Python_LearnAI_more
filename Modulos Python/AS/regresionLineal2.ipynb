```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.DataFrame()
data["Area"] = [2500, 3000, 3500, 4000, 4500]
data["Precio"] = [1550000, 2350000, 2500000, 3200000, 3535000]
```

```python
plt.scatter(data["Area"], data["Precio"])
```

```python
X = data[["Area"]]
```

```python
y = data[["Precio"]]
```

```python
modelo = linear_model.LinearRegression()
```

```python
modelo.fit(X, y)
```

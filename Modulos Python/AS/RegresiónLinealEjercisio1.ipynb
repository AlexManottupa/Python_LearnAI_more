```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.DataFrame()
df["Age"]=  [25, 30, 35, 40, 45]
df["Salary"] =  [3000, 3500, 4000, 4500, 5000]

```

```python
plt.scatter(df["Age"], df["Salary"], color="b", marker="o")
plt.title("Relacion entre edad y salario")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
```

```python
X = df[["Age"]]
```

```python
y = df[["Salary"]]
```

```python
modelo = linear_model.LinearRegression()
```

```python
modelo.fit(X, y)
```

```python
modelo.predict([[18]])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python

df = pd.read_excel('temeperaturas 2020 ala 24.xlsx')
df
```

```python
# Definir la variable objetivo y las características
df['compra'] = (df['Temp. Máx (°C)'] > 25).astype(int)  # Ejemplo de variable binaria
X = df[['Temp. Máx (°C)']]  # Característica
y = df['compra']  # Variable objetivo



# Crear el gráfico de dispersión
plt.scatter(df['Año'], df['Temp. Mín (°C)'], label='Temp. Mín (°C)', color='blue')
plt.scatter(df['Año'], df['Temp. Máx (°C)'], label='Temp. Máx (°C)', color='red')

# Añadir etiquetas y título
plt.xlabel('Año')
plt.ylabel('Temperatura (°C)')
plt.title('Temperaturas Mínimas y Máximas por Año')

# Añadir leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X, y, train_size=0.9, random_state=42)
```

```python
X_entrena
```

```python
modelo = LogisticRegression()
```

```python
modelo.fit(X_entrena, y_entrena)
```

```python
modelo.score(X_entrena, y_entrena)
```

```python
datos_nuevos = pd.DataFrame({'Temp. Máx (°C)': [20, 22, 27, 30]})
datos_nuevos
```

```python
probablididad = modelo.predict_proba(datos_nuevos)
probablididad
```
